## Background & Prior Work

Our topic started out with a general focus on housing attributes in certain neighborhoods. This idea evolved into us asking whether the quality of local schools have an influence on the safety of the surrounding area. Thus, we decided to dive deeper by investigating whether a school’s public ranking is related to the safety of its respective neighborhood. 

Schools are generally ranked based on standardized test scores and graduation rates [1]. However, other factors include average class size, facilities that endorse extracurricular activities, and school funding. The methods in measuring a US high school’s rank varies over time; however, recent rankings are consistently calculated using variables such as AP/IB exams, standardized tests, graduation rates, and state assessments [2]. 

Highly ranked schools are interpreted as providers of better education, which is often an important factor that families consider when buying houses. A study on the Fremont Unified School District in California reveals that higher school quality is associated with higher than average housing prices, suggesting that only families with higher income could afford to live in the area [3]. Conversely, the more funds a school has (derived from higher property taxes), the better the student outcomes, and consequently, the better the school’s ranking [4]. 

There have been many studies that explore the connection between schools and crime rates. For instance, a study from the U.S. Department of Education shows that there are higher crime rates in public high schools that are “located in areas with high social disorganization and residential mobility” [5]. Another study discovered a weak correlation between crime rates and the quality of high school education due to a large number of confounding variables [6]. Furthermore, an article that analyzed the impact of attending a first-choice high school found that an “increase in school quality leads to a reduction in the social cost of arrested crimes” [7]. Therefore, previous research on this topic reveals that there seems to be a relationship between crime rates and school quality. However, to our knowledge there has been no research on whether a public high school’s ranking correlates with the crime rate of a neighborhood, and how the income level of households factors into this relationship.

### School Rankings Data

To get the data for school rankings, we use the Greatschools.org website. Because there isn't a publicly available api to directly get the information for each of the schools, we need to scrape the website to get the information we need. When looking at the website at first, we see that all the rankings are organized in a tabular format, and we can use this to our advantage to scrape the rankings and other necessary information.

Because the entries in the tables are dynamically generated in JavaScript, we need to parse through the script tag in the HTML file format and find the pattern that is used for storing the contents in the tables.

```python
# contains the list of all the column names for our dataframe
school_ranking_features = ['name', 'rating', 'lat', 'lon', 'percentLowIncome', \
                           'districtName', 'districtCity', 'state', 'ZipCode']

rankings_data = pd.DataFrame(columns=school_ranking_features)
```

```python
# There are 125 pages of LA schools that have rankings and we notice that the url contains
# the information for the corresponding page number, so we use that to our advantage when scraping data.

count = 0
for pagenum in tqdm(range(1, 126)):
    url = 'https://www.greatschools.org/california/los-angeles/schools/?distance=60&gradeLevels%5B%5D=h&gradeLevels%5B%5D=e&gradeLevels%5B%5D=m&lat=34.0795&locationType=city&lon=-118.316&page={}&st%5B%5D=public&st%5B%5D=private&st%5B%5D=charter&view=list'.format(pagenum)
    
    # Parse the website in HTML format
    html_content = requests.get(url).content
    soup = bs(html_content, 'html.parser')
    
    out = soup.find('script').string

    # Extract the data contained in the tables
    d1 = out.split(';gon.')[9].replace('search=', '')
    
    # Convert the string to a json object (dictionary)
    my_dict = json.loads(d1)

    for i in my_dict['schools']:
        if i['rating'] is not None:
            count += 1
            
            # Load each school observation with specified variables to the dataframe
            rankings_data.loc[len(rankings_data)] = [i['name'], i['rating'], i['lat'], i['lon'], \
                               float(i['percentLowIncome'].replace('%', ''))/100, \
                               i['districtName'], i['districtCity'], i['state'], i['zipcode']]
            
rankings_data.head()
```

```python
# save dataframe to csv file
rankings_data.to_csv('total_la_school_rankings.csv', index=False)
```

Here is our final dataframe for school rankings in LA county. It has 3125 rows or schools from LA, each containing the name of the school, its rating on a scale of 1-10, latitude and longitude, percentage of low income students, district name, district city, state, and zip code.

### Combining School Rankings, Housing, and Crime Data

In this section, we will be combining the school rankings, housing, and crime data all into one dataset. We start by gathering all these datasets separately. We then combine the data using the Zip Code since this makes it specific enough to roughly fall into the same area and location, but not too specific like the latitude and longitude as these values may not be the exact same across the three datasets.

```python
# Store the dataset containing all the school rankings into a dataframe
df = pd.read_csv('total_la_school_rankings.csv')

# Store the LA housing data to a dataframe
ef = pd.read_csv('LAhousing.csv')

# Merge the school rankings data with the housing data based on Zip Code.
ff = pd.merge(df, ef[['ZipCode', 'TypicalHomeValue']], on='ZipCode').sort_values(by=['rating'], ascending=False)
ff.head()
```

```python
# Store the LA crime dataset to a dataframe
gf = pd.read_csv('la_crime_dataset.csv')

# Merge the Crime dataset with the School Rankings and Housing dataset.
merged_dataset = pd.merge(ff, gf[['ZipCode','crimes']].groupby(['ZipCode']).sum(), on='ZipCode', how='left').reset_index(drop=True)
merged_dataset = merged_dataset.fillna(0)

# Convert the entries in the crimes column to an integer
merged_dataset['crimes'] = merged_dataset['crimes'].map(int)

#show the dataset with all the entries where we have crime data
merged_dataset[merged_dataset['crimes'] != 0]
```

After combining all the datasets together, we end up with 1009 different observations and 11 different variables. They store important information such as the percentage of low income students and coordinates in addition to the rankings, crimes, and home values.

```python
# Store the merged LA dataset to a csv file
merged_dataset[merged_dataset['crimes'] != 0].to_csv('la_city_dataset_new.csv', index=False)
```

## Inferential Analysis - Linear Regression

### Ratings Predicting Crimes

```python
mod = smf.ols(formula='crimes ~ rating', data=df)
res = mod.fit()
print(res.summary())
```

School ratings predicts the variation in crimes by 11.7% (Adj. r-squared = 0.117), which suggests a relatively small correlation between school ratings and crimes.

The p-value is small which means this relationship is statistically significant. 

Therefore, we can reject the first part of our null hypothesis as school ratings are negatively correlated with their surrounding neighborhood's crime rates, albeit as a weak correlation. 

### Home Value Predicting Crimes

```python
mod = smf.ols(formula='crimes ~ log_TypicalHomeValue', data=df)
res = mod.fit()
print(res.summary())
```

Home values predicts the variation in crimes by 6.1% (Adj. r-squared = 0.061), which suggests a very weak correlation between home values and crimes.

The p-value is small which means this relationship is statistically significant. 

Therefore, we can reject the second part of our null hypothesis as home values are negatively correlated with their surrounding neighborhood's crime rates, albeit as a weak correlation. 

### Low Income Predicting Crimes

```python
mod = smf.ols(formula='crimes ~ percentLowIncome', data=df)
res = mod.fit()
print(res.summary())
```

The percentage of low-income students explains only about 16.5% of the variation in the number of crimes, as indicated by the adjusted R-squared value. The "Percent Low Income" variable is statistically significant due to its low p-value, indicating a strong association between the two variables.

### Ratings Predicting Home Value

```python
mod = smf.ols(formula='log_TypicalHomeValue ~ rating', data=df)
res = mod.fit()
print(res.summary())
```

Ratings predicts home values by 26.4% (Adj. r-squared = 0.264), which suggests a small/medium correlation between home values and ratings.

The p-value is small which means this relationship is statistically significant. 

Therefore, we can reject the second part of our null hypothesis as ratings are positively correlated with their surrounding neighborhood's home values. 

### Linear Regression Results Summary and Analysis

In summary, our linear regression models reveal that there indeed exists negative correlations between ratings and crimes and home values and crimes, and also confirms the positive correlation between low income percentage and crimes, and ratings and home value. All these relationships were statistically significant with small p-values. Looking at the adjusted R squares, 
- ratings predicts the variation in crimes by 11.7% 
- home values predicts the variation in crimes by 6.1%
- low income percentage predicts the variation in crimes by 16.5%
- ratings predicts home value by 26.4% 

Thus, our **null hypothesis is rejected and our alternative hypothesis is accepted** as there exists a weak negative correlation between LA school ratings and neighborhood crime rates, meaning that the higher the school's rating, the lower the crime rate in the neighborhood. Additionally, home value plays a role in this relationship by having a weak negative correlation with crime rates, and a weak/medium positive correlation with school ratings. In other words, as home value increases, crime rates decreases while ratings increases. 

We also discovered that low income percentage has a weak positive correlation with crime rates, meaning that as the percentage of low income students increases, crime rates increases as well. 

Next, we will take our project a step further and conduct Multiple Linear Regression to see whether multiple predictors could impact crime rates, as ratings, home value, and low income percentage might altogether explain crime rates better than as individual predictors. We will first look at whether ratings, home value, and percentage of low income students could predict crime rates; then, we will look at whether ratings and low income percentage could predict crime rates.

## Inferential Analysis - Multiple Linear Regression

### Ratings, Log-Transformed Home Value, and Low Income Predicting Crimes 

```python
mod = smf.ols(formula='crimes ~ percentLowIncome + log_TypicalHomeValue + rating', data=df)
res = mod.fit()
print(res.summary())
```

Ratings, home value, and percentage of low income students predicts crime rates by 17.0% (Adj. r-squared = 0.17), which suggests a weak correlation.

The p-value is small for both `percentLowIncome` and `rating`, which means these variables have statistically significant relationships with crime rates. However, home value has a high p-value (0.439), suggesting that even if it had a significant relationship with crimes by itself (as found in linear regression), it has a insignificant relationship compared to the other two predictors. Therefore, we will omit it from the next mulitple regression model and look at whether ratings and low income percentage could predict crimes better.   

### Ratings and Low Income Predicting Crimes 

```python
mod = smf.ols(formula='crimes ~ rating + percentLowIncome', data=df)
res = mod.fit()
print(res.summary())
```

17.0% (Adj. r-squared = 0.170) of the variation in crimes is explained by the predictors `rating` and `percentLowIncome`, suggesting a small correlation between these variables. 

The p-value for `rating` is 0.009 and the p-value for `percentLowIncome` is 0.000, which are both statistically significant.

However, the p-value for `percentLowIncome` is lower than that of `rating`, suggesting that `percentLowIncome` predicts crime rates better than `rating`. 

### Multiple Linear Regression Results Summary and Analysis

In summary, looking at the adjusted R squared values, our multiple linear regression models reveal that:
- percentage of low income students, home value, and ratings predicts the variation in crimes by 17.0%
   - with home value being insignificant compared to the other two predictors
- ratings and low income percentage predicts crimes by 17.0%

Unlike linear regression, we are unable to determine if our overall results are significant or not, especially for the three predictor model in which home value had a very large p-value.

To recap, our previous linear regression results were:
- ratings predicts the variation in crimes by 11.7% 
- home values predicts the variation in crimes by 6.1%
- low income percentage predicts the variation in crimes by 16.5%
- ratings predicts home value by 26.4% 

Therefore, it seems like our best model for predicting crimes could be either a combination of percentage of low income students, home value, and ratings, or ratings and low income percentage, as both predict a variation of 17.0%---the highest number out of all our models. This suggests that crimes rates are best explained by multiple variables instead of any individual variable. 

Low income percentage in particular is the strongest individual predictor, as its p-value is lower than rating and home value. Additionally, looking back at our results for linear regression, low income percentage alone predicts the variation in crimes by 16.5%, which is the next best model.

Meanwhile, home value is the weakest individual predictor, as it had a large p-value in multiple linear regression; this corresponds to our linear regression results in which home values was only able to predict 6.1% of the crimes, which is the weakest result out of all our models. 

# Ethics & Privacy

In other words, in order to remain ethical, our results must not be misconstrued as causal factors. We want to avoid highly biased and racist connotations being drawn from our analyses. Therefore, we made it clear in our analysis and in our conclusion that our results are entirely correlational, not causational.
